```python
import numpy as np
import pandas as pd
import tabula
```

```python
#Extraindo tabelas presente na pg 1 da PGDAS 
list_table = tabula.read_pdf("PGDAS.pdf", pages = "1", lattice = True)
table1 = list_table[0]
table2 = list_table[1]
```

```python
table2
```

```python
#Eliminando valores de receita do Mercado Externo
table2.drop(index = list(range(6,len(table2))), inplace = True)
#Adicionando Receita de 12/2021
table2.loc[5,'Unnamed: 5'] = '12/2020'
table2.loc[5,'Unnamed: 6'] = table1.loc[0, 'Mercado Interno']
```

```python
#Lista das colunas
cols = list(table2.columns)
#Nome das colunas que contêm apenas datas de competência
index = [col for i,col in enumerate(cols) if i%2 == 0]
#Nome das colunas que possuem os valores de receita
values = [col for i,col in enumerate(cols) if i%2 != 0]
```

```python
#Convertendo valores de receita para o formato americano
table2[values] = table2[values].applymap(lambda x: str(x).replace('.', '').replace(',', '.'))
#Convertendo strings para float
table2[values] = table2[values].applymap(lambda x: float(x))
table2
```

```python
#Series para os valores de receitas em cada um dos meses
pd.Series(
    #Tranformado ND arrays para 1D
    np.hstack(table2[values].values), 
    index = np.hstack(table2[index].values)
)
```
